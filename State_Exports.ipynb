```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file = "Exports_State.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
state_df = pd.read_csv(file)
for col in state_df:
    if col != 'State':
        state_df[col] = pd.to_numeric(pd.Series(state_df[col]), errors = 'coerce')
state_df.head()
```

```python
# Show just the header
top_five_df = state_df[1:6]

years = top_five_df.keys()
top_five_df

```

```python
Top5_df = top_five_df.T
Top5_df = Top5_df.reset_index()
Top5_df.columns = Top5_df.iloc[0]
Top5_df


cleaned_state_df = Top5_df.iloc[1:]
years = cleaned_state_df['State']

Texas = cleaned_state_df['Texas']
California = cleaned_state_df['California']
NY = cleaned_state_df['New York']
Washington = cleaned_state_df['Washington']
Illinois = cleaned_state_df['Illinois']
```

```python

# Plot the unemployment values for a single country
plt.plot(years, Texas, color="b", label="Texas")
plt.plot(years, California, color="r", label="California")
plt.plot(years, NY, color="g", label="New York")
plt.plot(years, Washington, color="k", label="Washington")
plt.plot(years, Illinois, color="orange", label="Illinois")

plt.xlabel("Years")
plt.ylabel("Revenues (100 billion)")
plt.title("Top 5 States exports revenue over 10 years")
```
