```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file = "Exports_State.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
state_df = pd.read_csv(file)
for col in state_df:
    if col != 'State':
        state_df[col] = pd.to_numeric(pd.Series(state_df[col]), errors = 'coerce')
state_df
```

```python
cleaned_state_df = state_df.T
cleaned_state_df = cleaned_state_df.reset_index()
cleaned_state_df.columns = cleaned_state_df.iloc[0]
cleaned_state_df


cleaned_state_df = cleaned_state_df.iloc[1:]
cleaned_state_df
```

```python
#set the years for x-axis
years = cleaned_state_df['State']

#data for the states
Texas = cleaned_state_df['Texas']
California = cleaned_state_df['California']
NY = cleaned_state_df['New York']
Washington = cleaned_state_df['Washington']
Illinois = cleaned_state_df['Illinois']
```

```python
# Plot the line graph for each states
plt.figure(figsize=(18,8))

plt.plot(years, Texas, color="b", label="Texas")
plt.plot(years, California, color="r", label="California")
plt.plot(years, NY, color="g", label="New York")
plt.plot(years, Washington, color="k", label="Washington")
plt.plot(years, Illinois, color="orange", label="Illinois")

plt.xlabel("Years",fontsize = 20 )
plt.ylabel("Revenues (100 billion)", fontsize = 20)
plt.title("Top 5 States exports revenue over 10 years", fontsize = 20)
plt.legend(title="States", loc= 9,bbox_to_anchor=(1.08, 1.02), fontsize = 15)
```
