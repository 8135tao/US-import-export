```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store filepath in a variable
file_one = "Commodity.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
file_one_df = pd.read_csv(file_one)
file_one_df.sort_values(by="2007 to 2017 (Dollar Change)")
```

```python
# Show just the header
sorted_df = file_one_df.sort_values(by="2007 to 2017 (Dollar Change)", ascending=False)
positive_states_df = sorted_df[0:5]
positive_states_df
#negative_states
sorted_ascending_df = file_one_df.sort_values(by="2007 to 2017 (Dollar Change)")
negative_states_df = sorted_ascending_df[0:5]
negative_states_df

```

```python

# Plot the unemployment values for a single country
plt.bar(positive_states_df['Item'], positive_states_df['2007 to 2017 (Dollar Change)'], color=['g'])

plt.ylabel("change over 10 years (%)")
plt.title("Top 5 states with most positive % change over 10 years")


x_axis = np.arange(len(positive_states_df))
tick_locations = [value+0.4 for value in x_axis]
plt.xticks(tick_locations, positive_states_df['Item'], rotation="vertical")
```

```python
plt.bar(negative_states_df['Item'], negative_states_df['2007 to 2017 (Dollar Change)'], color=['r'])
plt.ylabel("change over 10 years (%)")
plt.title("Top 5 states with most negative change over 10 years")
```
